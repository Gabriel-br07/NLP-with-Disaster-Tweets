```python


palavras_irrelevantes = nltk.corpus.stopwords.words("english")

    

def Limpar_Palavras_Irrelevantes(dados,coluna):
    frase_processada = []
    for opiniao in dados[coluna]:
        nova_frase = []
        palavras_texto = token_espaco.tokenize(opiniao)
        for palavras in palavras_texto:
            if palavras not in palavras_irrelevantes:
                nova_frase.append(palavras)
        frase_processada.append(" ".join(nova_frase))
    dados[coluna] =  frase_processada
    return frase_processada
    
```

```python
from string import punctuation

token_pontuacao = tokenize.WordPunctTokenizer()

pontuacao = []
for ponto in punctuation:
    pontuacao.append(ponto)
    
pontuacao_stopwords = palavras_irrelevantes + pontuacao
def Limpar_Pontuacao(dados,coluna,):
    frase_processada = []
    for opiniao in dados[coluna]:
        nova_frase = []
        palavras_texto = token_pontuacao.tokenize(opiniao)
        for palavras in palavras_texto:
            if palavras not in pontuacao_stopwords:
                nova_frase.append(palavras)
        frase_processada.append(" ".join(nova_frase))
    return frase_processada
```

```python
dados_treino["text"] = Limpar_Pontuacao(dados_treino,"text")
dados_treino.head()
```

```python
import unidecode

sem_acento = [unidecode.unidecode(texto) for texto in dados_treino["text"]]
stopwords_sem_acento = [unidecode.unidecode(texto) for texto in pontuacao_stopwords]
```

```python
def Limpar_acentos_misculo(dados,coluna):
    frase_processada = []
    for opiniao in dados[coluna]:
        nova_frase = []
        opiniao = opiniao.lower()
        palavras_texto = token_pontuacao.tokenize(opiniao)
        for palavras in palavras_texto:
            if palavras not in stopwords_sem_acento:
                nova_frase.append(palavras)
        frase_processada.append(" ".join(nova_frase))
    return frase_processada
```

```python
dados_treino["text"] = Limpar_acentos_misculo(dados_treino,"text")
```

```python

def stemmer(dados,coluna):
    stemmer = nltk.RSLPStemmer()
    frase_processada = []
    for opiniao in dados[coluna]:
        nova_frase = []
        opiniao = opiniao.lower()
        palavras_texto = token_pontuacao.tokenize(opiniao)
        for palavras in palavras_texto:
            if palavras not in stopwords_sem_acento:
                nova_frase.append(stemmer.stem(palavras))
        frase_processada.append(" ".join(nova_frase))
    return frase_processada

dados_treino["text"] = stemmer(dados_treino,"text")
```

```python
import spacy 
def lemma(dados,coluna):
    nlp = spacy.load("en_core_web_sm")
    frase_processada = []
    for opiniao in dados[coluna]:
        nova_frase = []
        opiniao = opiniao.lower()
        palavras_texto = token_pontuacao.tokenize(opiniao)
        for palavra in palavras_texto:
            token = nlp(palavra)
            if palavra not in stopwords_sem_acento:
                nova_frase.append(token[0].lemma_ if token else palavra)
        frase_processada.append(" ".join(nova_frase))
    return frase_processada

dados_treino["text"] = lemma(dados_treino,"text")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression 
tfidf= TfidfVectorizer(lowercase=False)
vetor_tfidf = tfidf.fit_transform(dados_treino["text"])
treino,teste,class_treino,class_teste = train_test_split(
                                        vetor_tfidf,dados_treino["target"],random_state=42)

Rl = LogisticRegression()
Rl.fit(treino,class_treino)
acuracia = Rl.score(teste,class_teste)
print(acuracia)
```
